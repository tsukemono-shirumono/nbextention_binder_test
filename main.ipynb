```python
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
```

```python
ts=np.arange(1000)/10

def f(a,b,delta,is_show_title,title):
    plt.plot(np.cos(a*ts),np.sin(b*ts+delta))
    if is_show_title:
        plt.xlabel(title)
    plt.show()

interact(f, a=2, b=2, delta=np.pi,is_show_title=True, title="sin ax")
```
